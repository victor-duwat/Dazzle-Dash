```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tmp = df.select_dtypes(include=['number']).corr ()
tmp.to_csv('data_corr.csv', index=False)
```

```python
tmp[((tmp <= -0.3) | (tmp >= 0.3))]
```

```python
# Create a scatterplot
sns.heatmap(data=tmp, annot=True)

# Show the plot
plt.show()
```

```python
life_expectancy_bmi = df.groupby(['Year'])[['Life expectancy ', ' BMI ']].mean()
life_expectancy_bmi.to_csv('life_expectancy_bmi.csv', index=True)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(life_expectancy_bmi.index, life_expectancy_bmi['Life expectancy '], label='Life Expectancy', color='blue', marker='o')

plt.plot(life_expectancy_bmi.index, life_expectancy_bmi[' BMI '], label='BMI', color='green', marker='o')

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Evolution of Life Expectancy and BMI over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

life_expectancy_bmi.plot(kind='bar', figsize=(10, 6), width=0.8)

plt.title("Life Expectancy and BMI by Year")
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='Life expectancy ', y=' BMI ', data=df, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title('Correlation with Regression Line between Life Expectancy and BMI')
plt.show()
```

```python
life_expectancy_adult_mortality = df.groupby(['Year'])[['Life expectancy ', 'Adult Mortality']].mean()
life_expectancy_adult_mortality.to_csv('life_expectancy_adult_mortality.csv')
```

```python
plt.figure(figsize=(10, 6))

plt.plot(life_expectancy_adult_mortality.index, life_expectancy_adult_mortality['Life expectancy '], label='Life Expectancy', color='blue', marker='o')

plt.plot(life_expectancy_adult_mortality.index, life_expectancy_adult_mortality['Adult Mortality'], label='Adult Mortality', color='green', marker='o')

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Evolution of Life Expectancy and Adult Mortality over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

life_expectancy_adult_mortality.plot(kind='bar', figsize=(10, 6), width=0.8)

plt.title("Life Expectancy and Adult Mortality by Year")
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='Life expectancy ', y='Adult Mortality', data=df, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title('Correlation with Regression Line between Life Expectancy and BMI')
plt.show()
```
